```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import random
import altair as alt
# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session



# upload dataset 
data= pd.read_csv("/kaggle/input/global-weather-repository/GlobalWeatherRepository.csv")
data.head()

random.seed(10)
df = data.sample(3500)

alt.Chart(df).mark_point(color='purple').encode(
    x='temperature_fahrenheit'
)

alt.Chart(df).mark_trail(color='purple').encode(
    x='air_quality_Carbon_Monoxide'
)

alt.Chart(df).mark_tick(color='purple').encode(
    x='air_quality_PM10'
)

c2=alt.Chart(df).mark_bar(color="purple").encode(
    y = "mean(wind_mph)",
    x = alt.X("condition_text", sort=alt.EncodingSortField(field='wind_mph', op='mean'))
    
).properties(title="Weather Condition Vs Average Wind (mph)",
    width=700,
    height=500
)
c2

alt.Chart(df).mark_bar(color="purple").encode(
    x=alt.X('humidity', bin=True),
    y='count()'
).properties(title="Humidity Bar Plot",
    width=700,
    height=500
)

interval = alt.selection_interval()
alt.Chart(df).mark_point().encode(
    x="visibility_miles",
    y="gust_mph",
    color=alt.condition(interval, 'gust_mph', alt.value('lightgray'),legend=None),
    tooltip=["visibility_miles", "gust_mph"]
).add_selection(
    interval
).properties(title="Visibility (miles) Vs Gust (mph)",
    width=700,
    height=500
)

df2 = df[(df["last_updated"].str.startswith('2023-08-29'))]

alt.Chart(df2).mark_bar(color='purple').encode(
    x='country',
    y='temperature_celsius',
    tooltip=["country","temperature_celsius"]
).properties(title='Temperature in Celcius across countries updated on 2023-08-29',
    width=900,
    height=300
)

alt.Chart(df2).mark_bar(color='purple').encode(
    x='country',
    y='air_quality_us-epa-index'
).properties(title='AQI across countries updated on 2023-08-29',
    width=900,
    height=300
)

cor_df = df[[ 'air_quality_Ozone', 
             'air_quality_PM10','air_quality_Carbon_Monoxide','air_quality_Nitrogen_dioxide',
             'air_quality_Sulphur_dioxide','air_quality_us-epa-index']]
corr_Mat = cor_df.corr()
 
corr_Mat = corr_Mat.round(2)
corr_Mat

interval = alt.selection_interval()
base=alt.Chart(df).mark_circle().encode(
    y="air_quality_us-epa-index", 
    color=alt.condition(interval, 'air_quality_PM10', alt.value('lightgray'),legend=None),
    tooltip=["air_quality_Carbon_Monoxide", "air_quality_PM10","air_quality_us-epa-index"]
    
).add_selection(
    interval
).properties(title='AQI Vs Carbon Monoxide and PM10',
    width=400,
    height=300
)

scatter = base.encode(x='air_quality_Carbon_Monoxide') | base.encode(x='air_quality_PM10')
scatter

# Store the SPLOM
#interval = alt.selection_interval()
chart = alt.Chart(df).mark_circle(color="purple",opacity=0.3).encode(
    alt.X(alt.repeat("column"), type="quantitative"),
    alt.Y(alt.repeat("row"), type="quantitative"),
    #color="condition_text",
    #color=alt.condition( interval,'condition_text', alt.value('lightgray'),legend=None),
    tooltip=["precip_in", "wind_mph","humidity","pressure_mb","gust_mph"]
).properties(
    width=125,
    height=125
).repeat( 
    row=["precip_in", "wind_mph", "humidity","pressure_mb","gust_mph"],
    column=["precip_in","wind_mph", "humidity","pressure_mb","gust_mph"]
).interactive()
chart
```
